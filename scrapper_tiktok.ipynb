# Scrapper TikTok

### In this notebook, I scraped all the items in the list of trending songs on TikTok.
### The website: https://ads.tiktok.com/business/creativecenter/inspiration/popular/music/pc/en

```python
# Import libraries
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By

import time
import pandas as pd

# Start Selenium driver
driver = webdriver.Firefox()

# Open TikTok website
driver.get("https://ads.tiktok.com/business/creativecenter/inspiration/popular/music/pc/en")

# Wait 4 seconds
time.sleep(4)

# Click in the menu of time
driver.find_element(By.XPATH,"/html/body/div[1]/div/div/div[1]/div[2]/div/div[1]/div[3]/div[1]/div[1]/span/span").click()

# Wait more 5 seconds
time.sleep(5)

# Selection "120 Last days" option
driver.find_element(By.XPATH, "/html/body/div[9]/div/div/div/div/div/div[3]").click()

# Wait more 5 seconds
time.sleep(5)

# Number of clicks in the button "View More"
num_clicks = 33

# Create a new dataframe
df = pd.DataFrame(columns=['Ranking','Songs', 'Artist', 'Url'])

# Loop of clicks
for songs in range(num_clicks):
    driver.find_element(By.XPATH, "/html/body/div[1]/div/div/div[1]/div[2]/div/div[3]/div/div[1]").click()
    time.sleep(4)
    
# Find the elements
    ranking = driver.find_elements(By.CSS_SELECTOR,'span.rankingIndex--CRstI')
    song = driver.find_elements(By.CSS_SELECTOR, 'span.music-name--Z2hNc')
    artist = driver.find_elements(By.CSS_SELECTOR, 'span.auther-name--3HglG')
    url = driver.find_elements(By.CSS_SELECTOR, '.sound-item-container--fNzli a')
    
#  Creating new lists, get the elements and add in the list

    list_ranking = []
    for num_ranking in ranking:
        list_ranking.append(num_ranking.text)
        
    list_songs = []
    for name_song in song:
        list_songs.append(name_song.text)
        
    list_artist = []
    for name_artist in artist:
        list_artist.append(name_artist.text)
        
    list_url = []
    for name_url in url:
        href = name_url.get_attribute('href')
        list_url.append(href)

# Combine lists of elements
    data_tuples = list(zip(list_ranking,list_songs,list_artist, list_url))
    
# Add the elements in the columns
    df = pd.DataFrame(data_tuples, columns=['Ranking','Songs', 'Artist', 'Url'])

    df.to_csv("trending_songs.csv", index=False)
```
